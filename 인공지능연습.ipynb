```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20, r=2):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)
    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much
    return divtime
plt.clf()
plt.imshow(mandelbrot(400, 400))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose
from tensorflow.keras.layers import Activation, BatchNormalization, Concatenate


""" Conv Block """
class ConvBlock(tf.keras.layers.Layer):
    def __init__(self, n_filters):
        super(ConvBlock, self).__init__()

        self.conv1 = Conv2D(n_filters, 3, padding='same')
        self.conv2 = Conv2D(n_filters, 3, padding='same')

        self.bn1 = BatchNormalization()
        self.bn2 = BatchNormalization()

        self.activation = Activation('relu')

    def call(self, inputs):
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = self.activation(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = self.activation(x)

        return x


""" Encoder Block """
class EncoderBlock(tf.keras.layers.Layer):
    def __init__(self, n_filters):
        super(EncoderBlock, self).__init__()

        self.conv_blk = ConvBlock(n_filters)
        self.pool = MaxPooling2D((2,2))

    def call(self, inputs):
        x = self.conv_blk(inputs)
        p = self.pool(x)
        return x, p


""" Decoder Block """
class DecoderBlock(tf.keras.layers.Layer):
    def __init__(self, n_filters):
        super(DecoderBlock, self).__init__()

        self.up = Conv2DTranspose(n_filters, (2,2), strides=2, padding='same')
        self.conv_blk = ConvBlock(n_filters)

    def call(self, inputs, skip):
        x = self.up(inputs)
        x = Concatenate()([x, skip])
        x = self.conv_blk(x)

        return x


""" U-Net Model """
class UNET(tf.keras.Model):
    def __init__(self, n_classes):
        super(UNET, self).__init__()

        # Encoder
        self.e1 = EncoderBlock(64)
        self.e2 = EncoderBlock(128)
        self.e3 = EncoderBlock(256)
        self.e4 = EncoderBlock(512)

        # Bridge
        self.b = ConvBlock(1024)

        # Decoder
        self.d1 = DecoderBlock(512)
        self.d2 = DecoderBlock(256)
        self.d3 = DecoderBlock(128)
        self.d4 = DecoderBlock(64)

        # Outputs
        if n_classes == 1:
            activation = 'sigmoid'
        else:
            activation = 'softmax'

        self.outputs = Conv2D(n_classes, 1, padding='same', activation=activation)

    def call(self, inputs):
        s1, p1 = self.e1(inputs)
        s2, p2 = self.e2(p1)
        s3, p3 = self.e3(p2)
        s4, p4 = self.e4(p3)

        b = self.b(p4)

        d1 = self.d1(b, s4)
        d2 = self.d2(d1, s3)
        d3 = self.d3(d2, s2)
        d4 = self.d4(d3, s1)

        outputs = self.outputs(d4)

        return outputs
```

```python
# 1. 필요한 라이브러리들을 불러옵니다.
import tensorflow as tf
from PIL import Image
import requests
import numpy as np

# 2. 미리 훈련된 EfficientNet-B0 모델을 불러옵니다.
# 'include_top=True'는 최종 분류 단계(1000개의 클래스 예측)까지 포함하겠다는 의미입니다.
# 'weights='imagenet''은 ImageNet 데이터셋으로 훈련된 가중치를 그대로 사용하겠다는 의미입니다.
model = tf.keras.applications.EfficientNetB0(
    include_top=True,
    weights='imagenet',
    classes=1000
)
print("EfficientNet-B0 모델을 성공적으로 불러왔습니다.")



# 이미지 파일 경로를 지정합니다. (예: C:/Users/MyUser/Desktop/cat.jpg)
# 파이썬 파일과 같은 폴더에 이미지를 두었다면 파일명만 적어도 됩니다.
image_path = '/content/drive/MyDrive/python-test-show/dog.png'

# 로컬 파일을 불러옵니다.
image = Image.open(image_path)
image = image.convert('RGB')
# EfficientNet-B0는 224x224 크기의 이미지를 입력으로 받으므로, 이미지 크기를 조절합니다.
image_resized = image.resize((224, 224), Image.LANCZOS)
print(f"이미지를 모델 입력 크기인 224x224로 변경했습니다.")

# (이후 4, 5, 6번 코드는 그대로 사용)

# EfficientNet-B0는 224x224 크기의 이미지를 입력으로 받으므로, 이미지 크기를 조절합니다.
# Image.LANCZOS는 이미지 크기를 줄일 때 화질 손실을 최소화하는 방법 중 하나입니다.


# 4. 이미지를 모델이 처리할 수 있는 형태로 변환합니다.
# 이미지를 numpy 배열(숫자 행렬)로 변환합니다.
image_array = np.array(image_resized)

# 모델은 한 번에 여러 장의 이미지를 처리할 수 있도록 설계되어 있습니다. (batch)
# 따라서 이미지 한 장을 예측하더라도, 묶음(batch) 형태로 만들어주어야 합니다.
# (1, 224, 224, 3) 형태로 차원을 확장합니다. (1장, 높이, 너비, 컬러채널)
image_batch = np.expand_dims(image_array, axis=0)

# EfficientNet 모델에 맞게 이미지 픽셀 값을 전처리합니다.
# 이 함수는 모델이 훈련될 때 사용했던 방식과 동일하게 값을 조정해줍니다.
processed_image = tf.keras.applications.efficientnet.preprocess_input(image_batch)

# 5. 모델을 사용하여 이미지를 예측(추론)합니다.
predictions = model.predict(processed_image)

# 6. 예측 결과를 사람이 보기 쉽게 해석하여 출력합니다.
# decode_predictions는 예측 결과를 (클래스 ID, 이름, 확률) 형태로 변환해줍니다.
# top=3는 가장 확률이 높은 상위 3개의 예측 결과만 보여달라는 의미입니다.
decoded_predictions = tf.keras.applications.efficientnet.decode_predictions(predictions, top=10)[0]

print("\n--- 예측 결과 ---")
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    # score는 0~1 사이의 확률 값이므로, 100을 곱해 퍼센트로 변환합니다.
    print(f"{i+1}순위: {label} ({score*100:.2f}%)")
```
